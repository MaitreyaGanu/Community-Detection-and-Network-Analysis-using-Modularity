This is due to the fact that the output is a two - dimesnional representation of a three - dimesnional graph. 

### 1 . Karate Club Graph 
The Karate Club Graph is a well-known network used in social network analysis, representing the relationships among 34 members of a karate club. This graph was first introduced in the study by Wayne W. Zachary in 1977. The dataset includes 34 nodes (representing the members of the club) and 78 edges (representing friendships between the members).

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community import modularity, greedy_modularity_communities

G = nx.karate_club_graph()

communities = list(greedy_modularity_communities(G))

modularity_value = modularity(G, communities)
print(f"Modularity: {modularity_value}")

community_colors = {}
for i, community in enumerate(communities):
    for node in community:
        community_colors[node] = i

plt.figure(figsize=(10, 8))

node_colors = [community_colors[node] for node in G.nodes]

nx.draw(G, node_color=node_colors, with_labels=True, cmap=plt.cm.rainbow, node_size=500, font_size=10)

plt.title("Graph with Community Detection (Modularity: {:.4f})".format(modularity_value))
plt.show()
```

### 2 . ErdsRnyi graph
An ErdsRnyi graph is a type of random graph where each pair of nodes (or points) in a graph has a fixed probability of being connected by an edge. Its like creating a network of nodes where each possible connection between nodes is randomly chosen based on a given probability

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community import modularity, greedy_modularity_communities

G = nx.erdos_renyi_graph(30, 0.1)

communities = list(greedy_modularity_communities(G))

modularity_value = modularity(G, communities)
print(f"Modularity: {modularity_value}")

community_colors = {}
for i, community in enumerate(communities):
    for node in community:
        community_colors[node] = i

node_colors = [community_colors[node] for node in G.nodes]
plt.figure(figsize=(8, 6))
nx.draw(G, node_color=node_colors, with_labels=True, cmap=plt.cm.rainbow, node_size=500, font_size=10)
plt.title(f"ErdsRnyi Graph (Modularity: {modularity_value:.4f})")
plt.show()
```

### 3 . BarabsiAlbert scale-free network
These networks are widely observed in real-world systems like the internet, social networks, and biological networks.

```python
G = nx.barabasi_albert_graph(30, 2)
communities = list(greedy_modularity_communities(G))


modularity_value = modularity(G, communities)
print(f"Modularity: {modularity_value}")


community_colors = {}
for i, community in enumerate(communities):
    for node in community:
        community_colors[node] = i

node_colors = [community_colors[node] for node in G.nodes]
plt.figure(figsize=(8, 6))
nx.draw(G, node_color=node_colors, with_labels=True, cmap=plt.cm.rainbow, node_size=500, font_size=10)
plt.title(f"BarabsiAlbert Network (Modularity: {modularity_value:.4f})")
plt.show()
```

### 4 . WattsStrogatz small-world network
It is a type of network where most nodes are not directly connected to each other, but can be reached through a small number of steps.

```python
G = nx.watts_strogatz_graph(30, 4, 0.1)

# Detect communities
communities = list(greedy_modularity_communities(G))

# Calculate modularity
modularity_value = modularity(G, communities)
print(f"Modularity: {modularity_value}")

# Visualization
community_colors = {}
for i, community in enumerate(communities):
    for node in community:
        community_colors[node] = i

node_colors = [community_colors[node] for node in G.nodes]
plt.figure(figsize=(8, 6))
nx.draw(G, node_color=node_colors, with_labels=True, cmap=plt.cm.rainbow, node_size=500, font_size=10)
plt.title(f"WattsStrogatz Small-World Network (Modularity: {modularity_value:.4f})")
plt.show()
```

### 5 . Les Misrables graph
The Les Misrables graph is a well-known network that represents the relationships between characters in Victor Hugo's novel Les Misrables.

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community import modularity, greedy_modularity_communities


G = nx.les_miserables_graph()


communities = list(greedy_modularity_communities(G))


modularity_value = modularity(G, communities)
print(f"Modularity: {modularity_value}")


community_colors = {}
for i, community in enumerate(communities):
    for node in community:
        community_colors[node] = i

node_colors = [community_colors[node] for node in G.nodes]
plt.figure(figsize=(10, 8))
nx.draw(G, node_color=node_colors, with_labels=False, cmap=plt.cm.rainbow, node_size=400)
plt.title(f"Les Misrables Graph (Modularity: {modularity_value:.4f})")
plt.show()
```
